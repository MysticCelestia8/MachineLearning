# Data Preprocessing

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Dataset

```
data_set = pd.read_csv("Covid_Data_new.csv")
X = data_set.iloc[:,:-1].values
Y = data_set.iloc[:,-1].values
```

```
print(X)
```

```
print(Y)
```




## Handling Missing Data

```
#replacing NaN values with mean values

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy = 'mean')
imputer.fit(X[:, 0:1])
X[:, 0:1] = imputer.transform(X[:, 0:1]) #similar to x = x+1
X[:, 0:1] = np.round(np.array(X[:, 0:1], dtype=float), decimals=2)

imputer.fit(X[:, 4:5])
X[:, 4:5] = imputer.transform(X[:, 4:5])
X[:, 4:5] = np.round(np.array(X[:, 4:5], dtype=float), decimals=2)
```

```
print(X)
```

## Encoding Categorical Data

### Encoding independent variables

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1,2,3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Encoding dependent variables

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```
print(Y)
```

## Splitting data into Test set & Training Set


```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=42)
```

```
print(X_train)
```

```
print (X_test)
```

```
print(Y_train)
```

```
print(Y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler();
X_train[:, 8:] = sc.fit_transform(X_train[:, 8:])
X_test[:, 8:] = sc.fit_transform(X_test[:, 8:])
```
