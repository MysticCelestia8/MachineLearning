```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x2 = [1,2,3,4,5]
y2 = [5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2)
```

```python

plt.plot(x,y,"y*")
plt.plot(x2,y2,"g--")
```

```python
x = np.array([1,2,3,4,5])
y = 2*(x**4) + 100

plt.plot(x,y,"b--")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = x**4
y2 = x**3
plt.plot(x,y,color='violet',linewidth=2)
plt.plot(x,y2,color='pink',linewidth=2)
```

```python
t = np.linspace(0, 2*np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)
plt.figure(figsize=(10, 8))
plt.plot(x, y, color='red', linewidth=2)
plt.show()
```

```python
profit = [10,20,30,40]
labels = ["Amazon","Flipkart","Snapdeal","Jiomart"]
colors = ["Blue","Orange","Red","Teal"]

plt.pie(profit,labels=labels,colors=colors,autopct="%.2f%%")
plt.show()
```
